### Removing twitter handles

```python
df['tweet_clean'] = df['tweet'].str.replace('@[\w]*', '')
df_test['tweet_clean'] = df_test['tweet'].str.replace('@[\w]*', '') 
df = df[['tweet','tweet_clean','label']]
df.head()
```

### Converting to lower case

```python
df['tweet_clean']= df['tweet_clean'].apply(lambda x: ' '.join(x.lower() for x in x.split()))
df_test['tweet_clean']= df_test['tweet_clean'].apply(lambda x: ' '.join(x.lower() for x in x.split()))
df.head()
```

### Numbers ain't required, let's remove them

```python
df['tweet_clean'] = df['tweet_clean'].str.replace('[0-9]', '')
df_test['tweet_clean'] = df_test['tweet_clean'].str.replace('[0-9]', '')

df.head()
```

### Removing URLs

```python
df['tweet_clean'] = df['tweet_clean'].str.replace('http\S+|www.\S+', '', case=False)
df_test['tweet_clean'] = df_test['tweet_clean'].str.replace('http\S+|www.\S+', '', case=False)
df.head()
```

### Removing Punctuations and Special Characters

```python
df['tweet_clean'] = df['tweet_clean'].str.replace('[^a-zA-Z ]', '')
df_test['tweet_clean'] = df_test['tweet_clean'].str.replace('[^a-zA-Z ]', '')
df.head()
```

### Removal of Stop words

```python
# Create stop words
stop = nltk.corpus.stopwords.words('english')
print(stop)
# Removal of stop words
df['tweet_clean'] = df['tweet_clean'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
df_test['tweet_clean'] = df_test['tweet_clean'].apply(lambda x: " ".join(x for x in x.split() if x not in stop))
df.head()
```

### Lemmatization
Lemmatization converts the word into its root word, rather than just stripping the suffices. It makes use of the vocabulary and does a morphological analysis to obtain the root word.

```python
nltk.download('wordnet')
```

```python
df['tweet_clean'] = df['tweet_clean'].apply(lambda x: ' '.join(tb.Word(word).lemmatize() for word in x.split()))
df_test['tweet_clean'] = df_test['tweet_clean'].apply(lambda x: ' '.join(tb.Word(word).lemmatize() for word in x.split()))
df.head()
```

### Extracting features from cleaned tweets

### Using TF-IDF weight
Term Frequency - Inverse Document Frequency

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer(min_df=79) # min_df =  1% of total observation

tfidf = vectorizer.fit_transform(df['tweet_clean'])
tfidf_test = vectorizer.transform(df_test['tweet_clean'])
```

```python
print(tfidf.shape)
print(tfidf_test.shape)
```

```python
tweets_tfidf= pd.DataFrame(tfidf.toarray(), columns=vectorizer.get_feature_names()) 
tweets_tfidf_test= pd.DataFrame(tfidf_test.toarray(), columns=vectorizer.get_feature_names()) 
```

```python
count_tfidf= tweets_tfidf[tweets_tfidf>0].count()
count_tfidf.sort_values(ascending=False)
```

## Data Preparation and Model Building
We'll check the accuracy of different models by splitting the train data and then for the most accurate model we'll fit the test data to upload the submission 

```python
# Assigning df['label'] as the dependent variable "y" and keeping "tweets_tfidf" the Independent variable "x"
y = df['label']
```

```python
# Splitting the data set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(tweets_tfidf, y, test_size = 0.3, random_state = 100)
```

### Using Logistic Regression

```python
# Using logistic regression
from sklearn.linear_model import LogisticRegression
logreg = LogisticRegression()
```

```python
# fitting the model
logreg.fit(x_train, y_train)
y_pred_logreg = logreg.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, f1_score

print('Confusion Matrix \n',  confusion_matrix(y_test, y_pred_logreg))
print('\n')
print('Classification Report: \n', classification_report(y_test, y_pred_logreg))
print('\n')
print('Accuracy: ', accuracy_score(y_test, y_pred_logreg)*100)
print('\n')
print('F1 Score: ', f1_score(y_test, y_pred_logreg))
```

### Using Support Vector Machine
Kernel = 'rbf'

```python
from sklearn.svm import SVC
svc_rbf = SVC(kernel='rbf')
svc_rbf.fit(x_train, y_train)
y_pred_svc= svc_rbf.predict(x_test)
```

```python
print('Confusion Matrix \n',  confusion_matrix(y_test, y_pred_svc))
print('\n')
print('Classification Report: \n', classification_report(y_test, y_pred_svc))
print('\n')
print('Accuracy: ', accuracy_score(y_test, y_pred_svc)*100)
print('\n')
print('F1 Score: ', f1_score(y_test, y_pred_svc))
```

Kernel = 'linear'

```python
svc_linear = SVC(kernel='linear')
svc_linear.fit(x_train, y_train)
y_pred_linear= svc_rbf.predict(x_test)
```

### Using Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(x_train, y_train)
```

```python
y_pred_rfc = rfc.predict(x_test)
print('Accuracy Score: ', accuracy_score(y_test, y_pred_rfc)*100)
print('F1 Score: ', f1_score(y_test, y_pred_rfc)) 
```

But this accuracy might lead to overfitting of the model. To avoid that we'll look for Cross-Validation for better results.

```python
# Using Cross-Validation for better results
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [300], 
    'max_features': range(50,145),
    'max_depth': range(5, 15, 5),
    'min_samples_leaf': range(50, 150, 50),
    'min_samples_split': range(50, 150, 50)
}
# Instantiate the grid search model
grid_search = GridSearchCV(estimator = rfc, param_grid = param_grid, 
                          cv = 5, n_jobs = -1,verbose = 1)
```

```python
grid_search.fit(x_train, y_train)
```

```python
# printing the optimal accuracy score and hyperparameters
print('We can get accuracy of',grid_search.best_score_,'using',grid_search.best_params_)
```

We can get accuracy of 74.25 using above grid parameters
